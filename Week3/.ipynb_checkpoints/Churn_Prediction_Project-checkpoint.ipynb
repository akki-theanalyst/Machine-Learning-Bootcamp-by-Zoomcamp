### Data preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Telecom_churn_csv data.csv')
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')

```

```python
df.columns
```

```python
categorical_col = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
categorical_col
```

```python
for col in categorical_col:

    df[col] = df[col].str.lower().str.replace(' ','_')
```

```python
df.head().T
```

```python
df.dtypes
```

```python
df['totalcharges']
```

```python
df['totalcharges'] = pd.to_numeric(df.totalcharges, errors = 'coerce')
```

```python
df[df['totalcharges'].isnull()][['customerid', 'totalcharges']]
```

```python
df.totalcharges = df.totalcharges.fillna(0)
```

```python
df.totalcharges.isna().sum()
```

```python
df.churn = (df.churn == 'yes').astype(int)
```

### Setting up the validation framework

**perform the train/validation/test split with scikit-learn**

```python
from sklearn.model_selection import train_test_split
```

```python
full_train_df, test_df = train_test_split(df, test_size = 0.2, random_state = 1)
```

```python
len(full_train_df), len(test_df)
```

```python
train_df, val_df = train_test_split(full_train_df, test_size = 0.25, random_state = 1)
```

```python
len(train_df), len(val_df), len(test_df)
```

```python
train_df
```

```python
train_df = train_df.reset_index(drop = True)
val_df = val_df.reset_index(drop = True)
test_df = test_df.reset_index(drop = True)
```
