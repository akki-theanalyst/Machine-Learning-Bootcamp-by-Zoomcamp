```python
X_train = X_train.reshape(6, 7150)
```

```python
y_train
```

```python
# Regression Model

def linear_regression(X, Y):

    ones = np.ones(X.shape[0])

    X = np.column_stack([ones, X])
    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX.inv.dot(X.T).dot(Y)
    

    return w_full[0], w_full[1:]
    
```

```python
b, w = linear_regression(X_train, y_train)  


```

```python
y_prediction = b + X_train.dot(w)
```

  ### RMSE (Root Mean Square Error)
  

```python
def rmse(y_pred, y):

    error = (y - y_pred)**2
    return np.sqrt(error.mean())
    
    
```
